```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("C:/Users/Orysya/Desktop/Data_Mining_III/Week7_Summary_Conclusion_Tips_HelpfulGuidance/forestfires.csv")
data
```

```python
def catFFMC(row):
    if row['FFMC'] < 70:
        return 'low'
    elif 70 < row['FFMC'] < 90:
        return 'high'
    else:
        return 'extreme'
data['catFFMC'] = data.apply(catFFMC, axis=1)
```

```python
def catDMC(row):
    if row['DMC'] < 30:
        return 'low'
    elif 30 < row['DMC'] < 40:
        return 'high'
    else:
        return 'extreme'
data['catDMC'] = data.apply(catDMC, axis=1)
```

```python
def catDC(row):
    if row['DC'] < 200:
        return 'low'
    elif 200 < row['DC'] < 300:
        return 'high'
    else:
        return 'extreme'
data['catDC'] = data.apply(catDC, axis=1)
```

```python
def catISI(row):
    if row['ISI'] < 10:
        return 'low'
    elif 10 < row['ISI'] < 16:
        return 'high'
    else:
        return 'extreme'
data['catISI'] = data.apply(catISI, axis=1)
```

```python
def newArea(row):
    return np.log(row['area'] + 1)
data['newArea'] = data.apply(newArea, axis=1)
```

```python
#Create variable dewpoint which will allow for variables: temp & RH be be removed
#using Td = T  - ((100 - RH)/5)
#note dewpoint is in C
def dewpoint(row):
    return row['temp'] - ((100 - row['RH'])/5)
data['dewpoint'] = data.apply(dewpoint, axis=1)
```

```python
#made binary value for rain
def newrain(row):
    if row['rain'] > 0:
        return 'yes'
    else:
        return 'no'
data['newrain'] = data.apply(newrain, axis=1)
```
